### Average NMSE over SNR in evaluation of 'ver4_' folder

```python
import matplotlib.pyplot as plt
import torch
import numpy as np

SNR = np.arange(0, 31, 5)

nmse_LS_LI = []
nmse_LS_NN = []
nmse_LI_NN = []

for snr in SNR:
    checkpoint = torch.load('static/CNN/BS16/3500_3516/ver4_/'+str(snr) +'dB/CNN_1_variable.pth')
    nmse_LS_LI.append(checkpoint['NMSE_LI'])
    nmse_LI_NN.append(checkpoint['NMSE_LI_NN'])
    nmse_LS_NN.append(checkpoint['NMSE_LS_NN'])

plt.figure(figsize=(10, 5))
plt.plot(SNR, nmse_LS_LI, label='LS+LI')
plt.plot(SNR, nmse_LS_NN, label='LS+NN')
plt.plot(SNR, nmse_LI_NN, label='LS+LI+NN')
plt.xlabel('SNR (dB)')
plt.ylabel('NMSE')
plt.title('Average NMSE over SNR in evaluation')
plt.legend()
plt.show()
```
